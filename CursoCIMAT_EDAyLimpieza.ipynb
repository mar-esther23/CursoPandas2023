Los datos son la primera versión de la transcripción de la Bitácora Onomástica del SEMEFO-DF de 1974, la cuál fue transcrita por personal de la CNB.

Quitar acentos, especiales y espacios múltiples

```python
def simplify_string(text, str_case='lower', remove_special=False):
    """Simplify a string to remove special characters, double spaces and use lower case."""
    if type(text)==str:
        if str_case=='lower':  text = text.lower()
        elif str_case=='upper':  text = text.upper()
        elif str_case=='title':  text = text.title()
        elif str_case=='capitalize':  text = text.capitalize()
        text = ' '.join([t for t in text.split()])
        text = unidecode(text)
        if remove_special:
            text = resub(r"[^a-zA-Z0-9]+", ' ', text)
        text = ' '.join([t for t in text.split()]) #paranoia
    return text

print( simplify_string(text) )
print( simplify_string(text, remove_special=True) )
```

```python
def to_dtype_with_errors(df, col, dtype='numeric', suffix='clean', replace_dict={'S-D':nan}, **kwargs):
    """
    Casts series astype and creates a column with succeses and an other with failures.
    
    Examples
    --------
    >>> df = to_dtype_with_errors(df, 'Fecha', dtype='datetime')
    >>> df = to_dtype_with_errors(df, 'Edad', dtype='numeric', downcast='integer')
    """
    # convert
    dat = df[col].replace(replace_dict)
    if dtype=='numeric':
        dat_clean = pd.to_numeric(dat, errors='coerce', **kwargs)
    elif dtype=='datetime':
        dat_clean = pd.to_datetime(dat, errors='coerce', **kwargs)
    else: return df
    # errors
    dat_error = dat.isna() != dat_clean.isna()
    dat_error = dat.where(dat_error)
    # insert after col
    index = df.columns.get_loc(col)
    df.insert(index+1, col+'_'+suffix, dat_clean)
    df.insert(index+2, col+'_error', dat_error)
    return df
```

```python
def replace_edades_menor_al_año(edad):
    from math import floor
    if type(edad)==str:
        edad_ = simplify_string(edad)
        edad_ = edad_.replace(' 1/2','.5')
        edad_ = edad_.replace('horas','hora').replace('dias','dia').replace('semanas','semana').replace('meses','mes').replace('anos','ano')
        dic_time = {'ano':1, 'mes':12, 'semana':52, 'dia':365, 'hora':8766}
        for nombre, n in dic_time.items():
            if nombre in edad_:
                edad_ = simplify_string( edad_.replace(nombre,'') )
                try: edad_ = float(edad_)
                except ValueError: return edad
                return floor( edad_ / n )
        dic_time = {'a':1, 'm':12, 'd':365}
        for nombre, n in dic_time.items():
            if nombre in edad_:
                edad_ = simplify_string( edad_.replace(nombre,'') )
                try: edad_ = float(edad_)
                except ValueError: return edad
                return floor( edad_ / n )
        try: edad_ = float(edad_)
        except(ValueError): return edad
        return floor(edad_)
    return edad

df['Edad_error'].apply(replace_edades_menor_al_año).unique()
```

```python
edad_replace = {'s-d':nan, '10 SEMNS':0, '1 Y 8 meses':0, '?':nan, '[8]':8}
df['Edad_estandar'] = df['Edad'].replace(edad_replace).apply(replace_edades_menor_al_año)
df['Edad_estandar'].unique()
```
